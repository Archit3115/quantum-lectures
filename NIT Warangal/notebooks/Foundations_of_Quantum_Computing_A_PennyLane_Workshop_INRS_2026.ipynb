## The Device-QNode architecture of PennyLane
In PennyLane, the hardware (or simulator) is decoupled from the circuit logic.
The `@qp.qnode(dev)` decorator turns a standard Python function into a quantum-classical hybrid object. Note that a QNode can also be instantiated directly as `qnode = qp.QNode(my_circuit, dev)`. See the [documentation on QNodes](https://docs.pennylane.ai/en/stable/code/api/pennylane.QNode.html) for more details.






## Quantum gates and observables

In PennyLane, quantum gates are operations. PennyLane distinguishes between (unitary) operations (gates like RX, CNOT) and observables (things you measure, like PauliZ - noting that the Pauli operations in particular (PauliX, PauliY and PauliZ) can be both observables and operations.

Note:

*   PennyLane provides broad support for single and multi-qubit gates. See the [documentation on quantum operators](https://docs.pennylane.ai/en/stable/introduction/operations.html) for examples.
*   Wires: In PennyLane, "wires" are the labels for qubits.
*   Composition: Gates are applied sequentially within the Python function.


## PennyLane's measurement suite

One of PennyLane's greatest strengths is its diverse measurement API. Note that, unlike real hardware, which would usually only provide samples, simulators can provide exact mathematical results. See the [Measurements documentation](https://docs.pennylane.ai/en/stable/introduction/measurements.html) for more details.

Some return types explained:

*   `qp.expval(O)`: Returns the expected value $\langle \psi | O | \psi \rangle$, where $O$ is some observable. Without specififying the number of samples ("shots"), this returns exact (analytic) results.
*   `qp.probs(wires)`: Returns a flat tensor of probabilities in lexicographical order of all the states (e.g., $[|00\rangle, |01\rangle, |10\rangle, |11\rangle]$ for a two-qubit circuit).
*   `qp.sample(O)`: Return raw samples from a single shot/execution.
*   `qp.counts(O)`: Return raw number of each result from a set number of shots.
*   `qp.state()`: Returns the full state vector. Note that this is only available to simulators.

##Extra material: General state preparation

State preparation is the process of mapping classical data (like a vector of numbers) into the amplitudes of a quantum state. While seeminly mathematically simple (apply a unitary $U$ such that $U|0\rangle = |\psi\rangle$) it is one of the most resource-intensive tasks in quantum computing. We saw some examples with the Bell state and GHZ state examples, but this is much more widely applicable in quantum algorithms. See the PennyLane documentation for [StatePrep](https://docs.pennylane.ai/en/stable/code/api/pennylane.StatePrep.html).

The fundamental challenge of state preparation is that a system of $n$ qubits has a Hilbert space of dimension $2^n$. To prepare a truly arbitrary state (where every amplitude is a unique complex number), you must provide $2^n$ pieces of information.

**Classical cost**: You must store $2^n$ complex numbers.

**Quantum cost**: You must apply a sequence of gates that correctly sets each of these amplitudes. Most generally, a common subroutine to do this is the Mottonen state preparation algorithm.
